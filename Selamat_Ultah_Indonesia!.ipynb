Step:


1.   Kita bakal buat di koordinat kartesius
2.   Buat lengkungan untuk bendera tiga, agar bisa diwarnai dua warna: merah dan putih
3.  Karena lengkungan bendera itu bentuk awalnya seperti grafik yang naik lalu turun, berarti kita perlu grafik yang akan naik di dua titik, misal x=1 dan x=2.
4. Karena turun di x=1 dan x=2, berarti grafik itu memiliki turunan sama dengan ketika x=1 dan x=2
5. Biar lebih mudah, kita anggap aja turunannnya itu
$(x-1)(x-2)\\
=x^2-3x+2\\
=0$
6. Karena turunannya adalah $y'=x^2-3x+2$, dengan menggunakan integral, diperoleh fungsinya adalah $y=\frac{1}{3}x^3-\frac{2}{3}x^2x+2x+c$
7. Karena kita butuh grafik, biar gampang, kita buat aja $c$ menjadi $c_1=0, c_2=2, c_3=4$.
Jadi, diperoleh:$Y_1 =\frac{1}{3}x^3-\frac{2}{3}x^2x+2x, Y_2 = \frac{1}{3}x^3-\frac{2}{3}x^2+2x+2, Y_3=\frac{1}{3}x^3-\frac{2}{3}x^2x+2x+4$.
8. Karena grafiknya naik dan turun di antara 1 dan 2, kita buat rangenya sekitar 0 sampai 2.5
9. Area antara $Y_1$ dan $Y_2$ menjadi warna putih dan area antara $Y_2$ dan $Y_3$ menjadi warna merah dari bendera
10. Tinggal $X=0$ jadi tiang benderanya!




```python
# Yok import dulu deh library yang kita butuhin gan,
# numpy buat titik-titiknya,
# matplotlib buat gambar benderanya,
# seaborn biar estetik dikit benderanya, nggak ancur-ancur amat
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Bikin tiga grafik Y_1 buat gambar lengkungan bendera paling bawah, Y_2 lengkungan tengah, Y_3 lengkungan paling atas, X_2 buat tiang bendera
X = np.linspace(0, 2.5, 400)
Y_1 = 1/3 * (X**3) - 3/2 * (X**2) + 2 * X
Y_2 = Y_1+2
Y_3 = Y_1 + 4
X_2 = np.array([0]*10000)
Y_5 = np.linspace(-10,min(Y_3),len(X_2))

# Biar figure gede dan background biru muda dikit
plt.figure(figsize=(8, 8))
plt.rcParams['axes.facecolor'] = '#d6f0ff'
plt.gcf().set_facecolor('#d6f0ff')

# yok kita warnain benderanya
sns.lineplot(x = X, y = Y_1, color='white')
sns.lineplot(x = X, y = Y_2, color='white')
sns.lineplot(x = X, y = Y_3, color='white')
plt.plot(X_2, Y_5, color ='black')
plt.fill_between(X, Y_1, Y_2, where=Y_1 < Y_2, color='white', alpha=1,
                 label=r'$\frac{1}{3}x^3 - \frac{3}{2}x^2 + 2x \leq \frac{1}{3}x^3 - \frac{3}{2}x^2 +2x, 0 \leq x \leq 2$')
plt.fill_between(X, Y_2, Y_3, where=Y_2 < Y_3, color='darkred', alpha=1,
                 label=r'$\frac{1}{3}x^3 - \frac{3}{2}x^2 + 2x + 2\leq\frac{1}{3}x^3 - \frac{3}{2}x^2 + 2x +4, 0 \leq x \leq 2$')

plt.xlabel(r'$X$', fontsize=10)
plt.ylabel(r'$Y$', fontsize=10)
# Kasih batasan untuk koordinat X
plt.xlim(-1,3)
# Kasih batasan untuk koordinat Y
plt.ylim(-6,6)
plt.legend(loc='lower right', fontsize = 10)
plt.grid(False)
plt.title('Lapor ketua! Bendera sudah ditemukan!', fontsize=13)


# Ilangin garis atas sama kanan biar nggak ganggu
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Yok tampilin benderanya
plt.show()
```
